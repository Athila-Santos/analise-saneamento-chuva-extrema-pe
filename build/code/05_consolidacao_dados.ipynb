```python
import pandas as pd
```

```python
# ---
# ETAPA 1: Carregar os 4 arquivos Parquet já processados
# ---
print("--- Iniciando Etapa 1: Carregando os datasets processados ---")

try:
    df_socio = pd.read_parquet('/content/drive/MyDrive/02 - Economia (UFPE)/2025.1/Tópicos de Macro (Economia do clima)/analise-saneamento-chuva-extrema-pe/build/output/pib_dens_hab_pe_2002_2021.parquet')
    df_saude = pd.read_parquet('/content/drive/MyDrive/02 - Economia (UFPE)/2025.1/Tópicos de Macro (Economia do clima)/analise-saneamento-chuva-extrema-pe/build/output/doencas_pe_2002_2021.parquet')
    df_clima = pd.read_parquet('/content/drive/MyDrive/02 - Economia (UFPE)/2025.1/Tópicos de Macro (Economia do clima)/analise-saneamento-chuva-extrema-pe/build/output/precipitacao_pe_2002_2021.parquet')
    df_saneamento = pd.read_parquet('/content/drive/MyDrive/02 - Economia (UFPE)/2025.1/Tópicos de Macro (Economia do clima)/analise-saneamento-chuva-extrema-pe/build/output/cobertura_agua_esgoto_pe_2002_2021.parquet')
    print("Todos os 4 arquivos Parquet foram carregados com sucesso.")
except FileNotFoundError as e:
    print(f"ERRO: Arquivo não encontrado. Verifique se todos os outputs da fase 'build' existem. Detalhe: {e}")
    # Adicione um 'exit()' aqui se estiver rodando como script.
```

```python
# ---
# ETAPA 2: A Grande Junção (Merge)
# ---
print("\n--- Iniciando Etapa 2: Unindo os datasets em um painel único ---")

# Começar com a base socioeconômica, que é a mais completa
painel_final = df_socio.copy()

# Juntar os outros DataFrames um a um
# Usamos 'how=left' para garantir que mantemos todos os municípios/anos da nossa base principal
painel_final = pd.merge(painel_final, df_saude, on=['cod_municipio', 'ano'], how='left')
painel_final = pd.merge(painel_final, df_clima, on=['cod_municipio', 'ano'], how='left')
painel_final = pd.merge(painel_final, df_saneamento, on=['cod_municipio', 'ano'], how='left')

# Verificar se a junção criou valores nulos (pode acontecer se houver alguma inconsistência)
print("\nVerificando valores nulos após o merge:")
print(painel_final.isnull().sum())
# Se houver muitos nulos, é um sinal para investigar a cobertura de dados de alguma das bases
```

```python
painel_final.to_parquet('/content/drive/MyDrive/02 - Economia (UFPE)/2025.1/Tópicos de Macro (Economia do clima)/analise-saneamento-chuva-extrema-pe/analysis/input/painel_final_pe.parquet', index=False)
```
